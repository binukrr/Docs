# Sets

```python
# Unique chars from list of chars
chars = ["a", "b", "b", "b", "c"]

unique_chars = set(chars)
print(unique_chars)
```

```python
# Unordered (index not possible)
unique_chars[0]
```

```python
# In
"b" in unique_chars
```

```python
# In
"c" in unique_chars
```

```python
# In
"d" in unique_chars
```

```python
# Methods: Add
unique_chars.add("d")
print(unique_chars)
```

```python
# Methods: Remove
unique_chars.remove("d")
print(unique_chars)
```

```python
# Lens chars vs unique
print(len(chars))
print((len)(unique_chars))
print ("Duplicates:", len(chars) - len(unique_chars))
```

```python
# Fun with sets: abcd & bdde
set1 = set(["a", "b", "c", "d"])
set2 = set(["b", "d", "d", "e"])
```

```python
# Intersection: set of identical items present in both sets
set1.intersection(set2)
```

```python
# Difference: items that are in set1 but not in set2
set1.difference(set2)
```

```python
# Union: all items from both sets (without duplicates)
set1.union(set2)
```

```python
# Symmetric difference: all items in either set1 or set2 but not both
set1.symmetric_difference(set2)
```

```python
# Issubset: True if all elements of set1 are in set2, otherwise false (eg:1)
set1.issubset(set2)
```

```python
# Issubset: True if all elements of set1 are in set2, otherwise false (eg:2)
set1 = set(["a", "b", "c", "d", "c"])
set2 = set(["b", "d", "d", "e", "a", "c"])
```

```python
set1.issubset(set2)
```
