# Importance of datatypes.

```python
# Combine int & str 
# will get error, see the error 'unsupported operand type(s) for +: 'int' and 'str''
age = 38
greeting = "hellow world"

print (age + greeting)
```

```python
# Combine strings - working example
greeting_1 = "hellow world! "
greeting_2 = "have a nice day."

print (greeting_1 + greeting_2)
```

```python
# String division
# will get an error since the division is unsupported for string and integer
greeting / 2
```

```python
# String division. Only the working way is :-
10 / 10
```

```python
# String multiplication
# It works as the multiplication.
age = 38
print (age * 2)
```

```python
# String multiplication
# With double quotes it takes as text 38.
age = "38"
print (age * 2)
```
